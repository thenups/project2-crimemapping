# Investigating Correlation: Economic Factors:

### Dependencies

```python
import pandas as pd
import numpy as np
from scipy.stats.stats import pearsonr 
```

### Import population data and create dataframe

```python
filepath = '../../raw/economic_stats/population_by_state_by_year.csv'
csv = pd.read_csv(filepath)
population_df = pd.DataFrame(csv)
```

```python
population_df.head()
```

### Import number of school shootings per year csv and create DataFrame

```python
filepath2 = '../../output/sparse_school_shootings_per_state_per_year.csv'
csv2 = pd.read_csv(filepath2)
shootings_df = pd.DataFrame(csv2)

# name year column (currently unnamed)
shootings_df = shootings_df.rename(columns={'Unnamed: 0': 'Year'})
```

```python
shootings_df.head(30)
```

```python
population_df.head(30)
```

### Import unemployment rate csv

```python
filepath3 = '../../raw/economic_stats/unemployment_by_state.csv'
csv3 = pd.read_csv(filepath3)
unemployment_rate_df = pd.DataFrame(csv3)
```

```python
unemployment_rate_df.head(30)
```

### Cut off population and shooting dataframes to 2015

```python
shootings_df = shootings_df[shootings_df['Year'] != 2018]
shootings_df = shootings_df[shootings_df['Year'] != 2017]
shootings_df = shootings_df[shootings_df['Year'] != 2016]

shootings_df = shootings_df.replace(np.nan, 0)

population_df = population_df[population_df['Year'] != 2018]
population_df = population_df[population_df['Year'] != 2017]
population_df = population_df[population_df['Year'] != 2016]

population_df = population_df.replace(np.nan, 0)

```

### Divide number of school shootings by population to get school shootings per capita

```python
shootings_percap_df = shootings_df / population_df
```

### Calculate correlation between a state's school shootings per capita and its unemployment rate

```python
np.corrcoef(shootings_percap_df['Illinois'], unemployment_rate_df['Illinois'])
```

```python
# Check correlations between school shootings per capita and unemployment rate
for element in population_df.columns.tolist()[1:]:
    print(element, 'Pearson correlation (25 yrs): ', pearsonr(shootings_percap_df[element], unemployment_rate_df[element]))
```
